```python
max_pop = df_ph["Nombre Habitants"].max()
max_dns = df_ph["Densité Population"].max()

fig = plt.figure(figsize=(15, 15))
ax = fig.gca()

for index, row in df_ph.iterrows():
    geoshape = json.loads(str(row["Geo Shape"]))
    shapecoords = geoshape["coordinates"]
    
    pop = float(row["Nombre Habitants"])
    dns = float(row["Densité Population"]) 
    
    color = (0, pop/(max_pop*1.5), pop/max_pop)
    BLUE = '#6699cc'
    
    ax.add_patch(PolygonPatch(geoshape, fc = color, ec = color, alpha = 0.5, zorder=4))
    ax.axis('scaled')

plt.show()
```
